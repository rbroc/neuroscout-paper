# New audio models

```python
mm = json.load(open('../models/tonal_rms.json'))
mm = {'tona': mm}
```

```python
lst = []
for k, v in mm['tona'].items():
    for t, a in v.items():
        lst.append(a)
```

```python
ids = lst
effects = ['tonal_centroid_' + str(i) for i in range(0,6)] + ['rms']
effects = {e: ids for e in effects}
process_set('audio_all', effects, download=True, exclude=['HealthyBrainNetwork'])
```
