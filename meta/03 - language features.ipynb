# Word frequency
Random-effects meta-analysis of the effects of word frequency in 9 datasets. Word frequency is operationalized as the Lg10WF measure from the SUBTLEX_US norms (Brysbaert & New, 2009). The model is identical in all datasets, and contains only log-transformed word frequencies and standard confound regressors (speech, motion, and temporal filtering).

```python
effects = {
    "subtlexusfrequencyLg10Wf": ["AqQxk", "Ma40P", "wjNOy", "9L2kY", "97pXv", "wkVjV", "M4jV2",
                                 "A25Pv", "wkV5V", "9BqG5"]
}
freq_map = process_set("word_freq", effects=effects, download=True, exclude=['SherlockMovie'], camel_case=False)
```

# Lexical effects — big model
An expanded model that includes 7 different lexical variables. This is meant to approximate the model reported in Yarkoni et al. (2008) for a reading paradigm, though the variables aren't identical, as some weren't available and others don't make sense in the context of auditory presentation.

```python
eff_names = ['massiveauditorylexicaldecisionNumphones', 'massiveauditorylexicaldecisionPhonlev',
             'subtlexusfrequencyLg10Wf', 'speech', 'concretenessConc.M', 'textLength',
             'massiveauditorylexicaldecisionDuration', 'massiveauditorylexicaldecisionNumsylls']
ids = ["AxBPb", "wJ3rm", "M37W8", "AYyjb", "9Bq37", "MD1B4", "wQDXz", "w0gXN", "wK2z5", "wQDpd"]
effects = {n: ids for n in eff_names}

ma_maps = process_set("lexical_effects", effects=effects, download=True, exclude=['SherlockMovie'], camel_case=False)
```
